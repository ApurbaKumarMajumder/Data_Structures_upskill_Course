Mutability and Hashability

* mutable objects are changable
* Immutable objects are not changable
* All immutable objects are hashable, but not all hashable objects are immutable.
* Hashtable is a feature of python objects that tells if the object has a hash value or not. It has a hash value that does not change during its entire lifetime.

### Str Format

```python
# string.format(var1, var2, .....)

print("{}, how are you?".format("Samantha"))
print("{}, {}, {}".format("one", 25, 3.5))
print("{2}, {0}, {1}".format(10, 20, 30))
```
