### Ways to define a Function

1. Takes Nothing, Returns Nothing
2. Takes Something, Returns Nothing
3. Takes Nothing, Return Something
4. Takes something, Return Something

```python
# Takes Nothing, Returns Nothing
def add():
    print("Enter two numbers ")
    a = int(input())
    b = int(input())
    ans = a + b
    print("Sum is", ans)

add()
```

```python
# Takes Something, Returns Nothing
def sub(a, b):
    ans = a - b
    print("Sub is", ans)

sub(10, 5)
```

```python
# Takes Nothing, Return Something
def add():
    print("Enter two numbers ")
    a = int(input())
    b = int(input())
    ans = a + b
    return ans

# print(add())
s = add()
print(s)
```

```python
# Takes something, Return Something
def add(a, b):
    ans = a + b
    return ans

s = add(10, 20)
print(s)
```

### Default Arguments

Default value indicated that the function argument will take that value if no argument value is passed during the function call.

The default value is assigned by using the assignment (=) operator.

Note: You cannot have non default argument after default argument.

```python
def add (a, b, c = 35, d = 40):
    ans = a + b + c + d
    return ans

print("Sum is", add(10, 20))
print("Sum is", add(10, 20, 100))
print("Sum is", add(100, 200, 300, 400))
```
