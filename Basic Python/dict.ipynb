# Dictionary

1. dict is a class
2. dict is mutable
3. dict is not hashable
4. dict is iterable
5. dict is not a sequence
6. dict cannot have duplicate keys (not data values)
7. indexing is not applicable to dict object
8. slicing operator is not applicable to dict object
9. dict elements are pair of key-value and data-value.

### How to create dict object

```python
# import image module
from IPython.display import Image
  
# get the image
Image(filename=r"C:\Users\9851a\Desktop\dict1.png")
```

```python
# dict is a function and therefore you will provide data in form of arguments (Keyword arguments)
d1 = dict(a = 10, b = 20, c = 30)
d1
```

```python
d1 = {102: 'Rahul', 105: 'Payal', 106: 'Arjun', 107:'Prachi'}
print(d1)
```

```python
print(d1[105])
```

```python
print(d1[107])
```

```python
# printing only values
for k in d1:
    print(d1[k])
```

```python
# printing both keys and values
for k in d1:
    print(k, ":", d1[k])
```

### How to edit dict elements?

Editing dict elements means you want to change data-value of the elements and not the key-value.

           dictObject[key-value] = newDataValue

```python
d1
```

```python
d1[102] = ('Naman', 'Dilip')
d1
```

```python
del d1[102]
```

### How to add new element in the dict Object?

        dictObject[new-key-value] = data-value

```python
d1[101] = 'Rahul'
d1
```

### Special Methods

All these methods are dict class attributes

1. items() -> collection dict elements
2. keys() -> collection of keys only of the elements
3. values() -> collection of data-values only of the dict elements

```python
d1.items()
```

```python
type(d1.items())
```

```python
for i in d1.items():
    print(i)
```

```python
# unpacking of the elements
for i, j in d1.items():
    print(i, j)
```

```python
d1.values()
```
