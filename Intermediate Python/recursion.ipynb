# Recursion

Function calling itself is called recursion.

```python
def f1(n):
    if n == 1:
        return 1
    s = n + f1(n-1) # f1 calls itself 
    return s

r = f1(3)
print(r) 
```

### How to approach recursive Function?

```python
# Write a recursive function to calculate sum of first n natural numbers.

''' 
1. add(n) is already created
    1 + 2 + 3 + .... + n
2. Recursive case
    n + add(n-1)
3. Base Case
    n == 1
    return 1
'''

def add(n):
    if n == 1:
        return 1
    
    return n + add(n-1)

print(add(10))
```

```python
# Write a recursive function to calculate sum of square of first n natural numbers.

'''
1. SquareAdd(n^2) is already created
2. Recursive case
    n ** 2 + squareAdd((n-1)**2)
3. Base case
    n == 1
    return 1**2
'''

def squareAdd(n):
    if n == 1:
        return 1**2
    
    return n**2 + squareAdd(n-1)

print(squareAdd(4))
```

```python
# Write a recursive function to print first n natural numbers.

def printN(n):
    if (n > 0):
        printN(n-1)
        print(n)

printN(10)
```
