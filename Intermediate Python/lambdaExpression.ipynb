# Lambda Expression

1. lambda is a keyword
2. lambda expressions are syntactically restricted to a single expression.
3. no need to def keyword

        lambda input: expression
        lambda a, b: a + b

```python
# both the functions are same
# does the same job
def add(a, b):
    return a+b

lambda a, b: a+b
```

### Recursion using lambda

```python
# Calculate factorial of a number using recursion lambda expression

f = (lambda a: 1 if a == 0 else a * f(a-1))
f(1111)
```
