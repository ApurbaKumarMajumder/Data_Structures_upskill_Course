### StopIterator

When next method is called for an iterator object which surpasses the last element of the containers object, it produces <b> StopIterator </b> expression.

* iterator object can only access elements in a sequence from first element to the last elements.

* next method is used to advances the iterator object.

* there is no previous method to traverse iterator object on the element in reverse direction.
Similarly, you cannot randomly access the elements using iterator.

```python
t1 = (5, 6, 4, 2, 8, 1, 9, 0)
it = iter(t1)

while True:
    try:
        print(next(it), end = " ")
    except StopIteration:
        break

print("Last line of the program")
```

Iterator object can be obtained in two ways

    1. using iter() method
    2. generator function

### Generator

* Generators are special kind of functions.

* A generator-function is defined like a normal function, but whenever it needs to generate a value, it does so with the yeild keyword rather than return.

* if the body of a def contains yeild, the function automatically becomes a generator function.

```python
def f1():
    yield 10
    print("one")
    yield 20
    print("Two")
    yield 30
    print("Three")

for e in f1():
    print(e)

it = f1()
```

```python
next(it)
```

```python
# Create a generator to produce first n even natural numbers.

def evenGenerator(n):
    a = 2
    while n:
        yield a
        a += 2
        n -= 1
```

```python
for e in evenGenerator(10):
    print(e, end = " ")
```

```python
for e in evenGenerator(int(input("Enter a number "))):
    print(e, end = " ")
```

### Endless iterator

```python
# We can create a generator object, which is capable of producing endless iterator.

# ex: fibonacci series

def generateFib(n):
    a, b = 0, 1
    while n:
        yield a
        a, b = b, a+b
        n -= 1
```

```python
for e in generateFib(int(input("Enter a number "))):
    print(e, end = " ")
```

```python
l1 = [e for e in generateFib(int(input("Enter a number ")))]
l1
```

```python
# endless iterator

def generateFib():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a+b

it = generateFib()
fib_list = []
```

```python
while True:
    print("Do you want to generate another element [y/n]")
    ans = input()
    if ans == 'y':
        x = next(it)
        fib_list.append(x)
        print(x)
    else:
        break
```

```python
fib_list
```

The return statement in generator is equivalent to raise StopIteration.
